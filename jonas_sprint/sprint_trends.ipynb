## DATA SPRINT
#### Strength of U.S. Market

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
energymkt = pd.read_csv('./datasets/owid-energy-data.csv')
```

```python
energymkt.shape
```

```python
pd.set_option('display.max_column', None)
energymkt.head(1)
```

```python
energymkt.columns
```

```python
usmkt = energymkt[energymkt['iso_code'] == 'USA']
```

```python
windmkt = energymkt[['year', 'iso_code', 'population', 'gdp', 'wind_cons_change_pct', 'wind_cons_change_twh', 'wind_consumption', 'wind_elec_per_capita', 'wind_electricity', 'wind_energy_per_capita', 'wind_share_elec','wind_share_energy']]
```

```python
windmkt
```

```python
uswindmkt = windmkt[windmkt['iso_code'] == 'USA']
```

```python
uswindmkt
```

```python
x=usmkt['year']
y=usmkt['wind_share_energy']

plt.plot(x, y)

plt.title('Share of Electricity Generation by Wind over Time')
plt.xlabel('Year')
plt.ylabel('% of Electricity Generation')
plt.show()
```

*above graph shows that wind is still a tiny portion of the overall energy market, but one that is rapidly growing proportionally! Shows ~4x increase from 2010 to 2022.*

```python
y1=usmkt['solar_share_energy']
y2=usmkt['hydro_share_energy']
y3=usmkt['other_renewables_share_energy']

plt.plot(x, y, label='wind')
plt.plot(x, y1, label='solar')
plt.plot(x, y2, label='hydro')
plt.plot(x, y3, label='other')

plt.title('Share of Electricity Generation by Renewable Energy Sources over Time')
plt.xlabel('Year')
plt.ylabel('% of Electricity Generation')
plt.legend()
plt.show()
```

*both wind and solar look like promising markets based on recent growth! and wind is now the top renewable energy market in 2020!*
