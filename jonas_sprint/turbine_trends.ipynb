## DATA SPRINT
#### Trends over Time

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wind = pd.read_csv('./datasets/wind_turbine_market.csv')
```

```python
wind.shape
```

```python
pd.set_option('display.max_column', None)
wind.head(2)
```

```python
wind.drop(columns=['faa_ors', 'faa_asn', 'usgs_pr_id', 't_img_srce', 't_img_date'], inplace=True)
```

```python
cap_growth = wind[['p_year', 't_cap']].groupby(by='p_year', as_index=False).sum()
cap_growth
```

```python
# To show overall growth of wind energy production by year

x = cap_growth['p_year']
y = cap_growth['t_cap']

plt.plot(x, y)
plt.title('Sum of Total Capacity Built Each Year')
plt.xlabel('Year')
plt.ylabel('Sum of Capacity (kW)')
plt.show()
```

```python
t_size_growth = wind[['p_year', 't_hh', 't_rd', 't_rsa', 't_ttlh']].groupby(by='p_year', as_index=False).mean()
t_size_growth
```

```python
# to show avg turbine size over time

x = t_size_growth['p_year']
y1 = t_size_growth['t_hh']
y2 = t_size_growth['t_rd']
y3 = t_size_growth['t_rsa']
y4 = t_size_growth['t_ttlh']


fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Hub Height (m)')
axs[0, 1].plot(x, y2, 'tab:orange')
axs[0, 1].set_title('Rotor Diameter (m)')
axs[1, 0].plot(x, y3, 'tab:green')
axs[1, 0].set_title('Rotor Swept Area (m2)')
axs[1, 1].plot(x, y4, 'tab:red')
axs[1, 1].set_title('Total Height (m)')


for ax in axs.flat:
    ax.set(xlabel='year')

for ax in axs.flat:
    ax.label_outer()
```

```python
pd.set_option('display.max_row', None)
top_manu = wind['t_manu'].value_counts()
top_manu
```

```python
topwind = wind.apply(lambda row: row[wind['t_manu'].isin(['GE Wind', 'Vestas', 'Siemens', 'Mitsubishi', 'Gamesa', 'Siemens Gamesa Renewable Energy'])])
topwind_growth = topwind.groupby(by='p_year')
topwind_growth.head()
```

```python
# to view top manufacturers growth over time
vestas = topwind[topwind['t_manu'] == 'Vestas']
ge_wind = topwind[topwind['t_manu'] == 'GE Wind']
siemens = topwind[topwind['t_manu'] == 'Siemens']
mitsubishi = topwind[topwind['t_manu'] == 'Mitsubishi']
gamesa = topwind[topwind['t_manu'] == 'Gamesa']
siemens_gamesa = topwind[topwind['t_manu'] == 'Siemens Gamesa Renewable Energy']

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(12,4), sharey=True, sharex=True)

fig.set_facecolor('lightsteelblue') # addressing plt.figure

ax[0, 0].plot(vestas['p_year'], vestas['p_tnum'], color='c')
ax[0, 0].set_title('Vestas', fontsize=10)
ax[0, 1].plot(ge_wind['p_year'], ge_wind['p_tnum'], color='m')
ax[0, 1].set_title('GE Wind', fontsize=10)
ax[0, 2].plot(siemens['p_year'], siemens['p_tnum'], color='y')
ax[0, 2].set_title('Siemens', fontsize=10)
ax[1, 0].plot(mitsubishi['p_year'], mitsubishi['p_tnum'], color='c')
ax[1, 0].set_title('Mitsubishi', fontsize=10)
ax[1, 1].plot(gamesa['p_year'], gamesa['p_tnum'], color='m')
ax[1, 1].set_title('Gamesa', fontsize=10)
ax[1, 2].plot(siemens_gamesa['p_year'], siemens_gamesa['p_tnum'], color='y')
ax[1, 2].set_title('SGRE', fontsize=10)


# manu_names = ['Vestas', 'GE Wind', 'Siemens', 'Mitsubishi', 'Gamesa', 'Siemens Gamesa Renewable Energy']

# for i in range(len(ax)):
#     ax[i, i].set_xlabel('Year', fontsize=12)
#     ax[i, i].set_title(f'{manu_names[i]}', fontsize=16)
    
# ax.set_ylabel('Number of Turbines', fontsize=12)

fig.suptitle('Growth of Turbines by Manufacturers over Time', fontsize=14)
plt.tight_layout()
```
